```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, plot, iplot

init_notebook_mode(connected=True)
```

```python
airpassengers = pd.read_csv('../../data/airpassengers.csv', index_col=0)

airpassengers.describe()
```

```python
airpassengers['log'] = airpassengers['x'].apply(np.log)

airpassengers['diff'] = airpassengers[['log']].rolling(13).apply(lambda x: x[-1] - x[0])

airpassengers['dates'] = pd.Series(pd.date_range(start='1949-1-1', periods=len(airpassengers.index), freq='M'), index = airpassengers.index)
```

```python
data = []

for col_name in ['x', 'log', 'diff']:
    trace = go.Scatter(
        x = airpassengers['dates'],
        y = airpassengers[col_name],
        mode = 'lines',
        name = col_name + ' vs Date'
    )
    
    data.append(trace)
    
iplot(data, show_link=False)

plot(data, filename='plot.html', auto_open=False);
```
