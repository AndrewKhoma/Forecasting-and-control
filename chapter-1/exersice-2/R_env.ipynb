```r
library(datasets)

str(AirPassengers)

write.csv(AirPassengers, file = "../../data/airpassengers.csv")
```

```r
df <- as.data.frame(AirPassengers)

df$log <- log(AirPassengers)

str(df)

df_diff <- diff(x = AirPassengers, differences = 1, lag = 12)

str(df_diff)
```

```r
plot(df$x, ylab = 'Passanger number in thousands', main = 'Without log transformation')
```

```r
plot(df$log, ylab = 'Passanger number in thousands', main = 'Log transformation')
```

```r
plot(df_diff, ylab = 'Passanger number in thousands', main = 'First order diff transformation, lag = 12')
```
