```r
library(datasets)

str(airquality)

write.csv(airquality, file = "../../data/airquality.csv")
```

```r
df <- as.data.frame(airquality)

# df[is.na(df)] <- 0
# df <- na.omit(df)

df$Fahrenheit <- df$Temp

df$Celsius <- (df$Fahrenheit - 32)*5/9

df$Date <- as.Date(paste(df$Day, df$Month, "1973", sep="-"), "%d-%m-%Y")

df <- df[, c("Ozone", "Solar.R", "Wind", "Fahrenheit", "Celsius", "Date")]

str(df)
```

```r
plot(df$Date, df$Ozone, type="l", main="Date vs Ozone", xlab="Date", ylab="Ozone")
```

```r
plot(df$Date, df$Solar.R, type="l", main="Date vs Solar radiation", xlab="Date", ylab="solar radiation")
```

```r
plot(df$Date, df$Wind, type="l", main="Date vs Wind", xlab="Date", ylab="Wind")
```

```r
plot(df$Date, df$Fahrenheit, type="l", main="Date vs Temp (Fahrenheit)", xlab="Date", ylab="Temp")
```

```r
plot(df$Date, df$Celsius, type="l", main="Date vs Temp (Celsius)", xlab="Date", ylab="Temp")
```
