```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import plotly

import plotly.tools as tls
import plotly.plotly as py
import plotly.graph_objs as go
import plotly.figure_factory as ff
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

init_notebook_mode(connected=True)
```

```python
airquality = pd.read_csv('../../data/airquality.csv', index_col=0)

airquality.describe()
```

```python
airquality['Year'] = 1973

airquality['Date'] = pd.to_datetime(airquality[['Year', 'Month', 'Day']])

airquality = airquality.rename(index=str, columns={'Temp' : 'Fahrenheit'})

airquality['Celsius'] = (airquality['Fahrenheit'] - 32) * float(5)/float(9)

airquality = airquality.drop(columns=['Day', 'Month', 'Year'])

airquality.head()
```

```python
data = []

for col_name in ['Ozone', 'Solar.R', 'Wind', 'Celsius', 'Fahrenheit']:
    trace = go.Scatter(
        x = airquality['Date'],
        y = airquality[col_name],
        mode = 'lines',
        name = col_name + 'vs Date'
    )
    
    data.append(trace)
    
iplot(data, show_link=False)
```
