```python
import time
import matplotlib
import numpy as np
import pandas as pd
from tqdm import tqdm

import matplotlib.pyplot as plt
%matplotlib inline

from datetime import datetime
import pandas_datareader as pd_dr
```

```python
tickers_name = []

with open('stocks_tickers_600_all.txt', 'r') as f:
    read_data = f.readlines()
    for line in read_data:
        tickers_name.append(line.split("\t")[0])
        
with open('mega_tickers_500_all.txt', 'r') as f:
    read_data = f.readlines()
    for line in read_data:
        tickers_name.append(line.split("\t")[0])

with open('all_stocks_tickers.txt', 'w') as f:
    f.write(' '.join(tickers_name))
```

```python
stock_df = pd.DataFrame(columns=['ticker', 'open', 'close', 'volume'])

stock_names = tqdm(tickers_name)
for stock in stock_names:
    try:
        stock_exchange = pd_dr.data.get_data_yahoo(stock, start='2019-01-02', end='2019-03-14')
        count_cur = stock_exchange.describe()[['Open', 'Close', 'Volume']].loc[['count']].values.flatten()
        count_cur = np.sort(count_cur)
        if count_cur[0] == count_cur[-1]:
            cur_stock = {'ticker': stock,
                         'open': count_cur[0], 
                         'close': count_cur[1], 
                         'volume': count_cur[2]}
            stock_df = stock_df.append(pd.Series(cur_stock), ignore_index = True)
    except Exception as e: 
        print("Something went wrong " + stock)
        print(e)

    stock_names.set_description("Processing %s" % stock)

stock_df.to_csv('stocks_count.csv')
```
